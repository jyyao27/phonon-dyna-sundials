```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.colors import to_rgb

def draw_BZ_border(ax, BZ_array, order_dir=None, **kwargs):

    """
    plot the border of the first Brillouin zone.
    """
    if order_dir is not None:
      
        for i in range(BZ_array.shape[0]-1):
         
            # foreground
            zf = 10
            lsf = '-'

            # background
            zb = 0
            lsb = '--'
         
            if order_dir.shape[0] == 3:
                dirvec = order_dir/np.linalg.norm(order_dir)
            elif order_dir.shape[0] == 2:
                azim = order_dir[0]/180.0*np.pi
                elev = order_dir[1]/180.0*np.pi
                r = 1.0

                x = r * np.cos(elev) * np.cos(azim)
                y = r * np.cos(elev) * np.sin(azim)
                z = r * np.sin(elev)

                dirvec = np.array([x,y,z])

            else:
                sys.exit('draw_BZ_border: Wrong order_dir')

            lim = -0.4

            r1 = BZ_array[i,:]
            r2 = BZ_array[i+1,:]

            p1 = np.dot(dirvec,r1)
            p2 = np.dot(dirvec,r2)

            if p1 >= lim and p2 >= lim:
                zorder = zf
                ls = lsf
                c = 'red'
            else:
                zorder = zb
                ls = lsb
                c = 'blue'

            ax.plot( [r1[0],r2[0]], [r1[1],r2[1]], [r1[2],r2[2]], zorder=zorder, ls = ls,**kwargs )

    else:
        ax.plot(BZ_array[:,0],BZ_array[:,1],BZ_array[:,2],**kwargs)



def plot_border(fig, ax, kpts, bg):
    """
    plot the border of the first Brillouin zone.
    """
   
    BZ_array = np.array([[ 0.25, -0.25, -0.5 ],
          [-0.25, -0.5 , -0.75],
          [-0.5 , -0.25, -0.75],
          [-0.25,  0.25, -0.5 ],
          [ 0.25,  0.5 , -0.25],
          [ 0.5 ,  0.25, -0.25],
          [ 0.25, -0.25, -0.5 ],
          [ 0.25, -0.5 , -0.25],
          [-0.25, -0.75, -0.5 ],
          [-0.25, -0.5 , -0.75],
          [-0.25, -0.75, -0.5 ],
          [-0.5 , -0.75, -0.25],
          [-0.25, -0.5 ,  0.25],
          [ 0.25, -0.25,  0.5 ],
          [ 0.5 , -0.25,  0.25],
          [ 0.25, -0.5 , -0.25],
          [ 0.5 , -0.25,  0.25],
          [ 0.75,  0.25,  0.5 ],
          [ 0.75,  0.5 ,  0.25],
          [ 0.5 ,  0.25, -0.25],
          [ 0.75,  0.5 ,  0.25],
          [ 0.5 ,  0.75,  0.25],
          [ 0.25,  0.5 , -0.25],
          [ 0.5 ,  0.75,  0.25],
          [ 0.25,  0.75,  0.5 ],
          [-0.25,  0.5 ,  0.25],
          [-0.5 ,  0.25, -0.25],
          [-0.25,  0.25, -0.5 ],
          [-0.5 ,  0.25, -0.25],
          [-0.75, -0.25, -0.5 ],
          [-0.5 , -0.25, -0.75],
          [-0.75, -0.25, -0.5 ],
          [-0.75, -0.5 , -0.25],
          [-0.5 , -0.75, -0.25],
          [-0.25, -0.5 ,  0.25],
          [-0.5 , -0.25,  0.25],
          [-0.75, -0.5 , -0.25],
          [-0.5 , -0.25,  0.25],
          [-0.25,  0.25,  0.5 ],
          [ 0.25,  0.5 ,  0.75],
          [ 0.5 ,  0.25,  0.75],
          [ 0.25, -0.25,  0.5 ],
          [ 0.5 , -0.25,  0.25],
          [ 0.75,  0.25,  0.5 ],
          [ 0.5 ,  0.25,  0.75],
          [ 0.25,  0.5 ,  0.75],
          [ 0.25,  0.75,  0.5 ],
          [-0.25,  0.5 ,  0.25],
          [-0.25,  0.25,  0.5 ]])
   

    BZ_array = np.matmul(BZ_array, bg)


    plot_type = 'border'
    ax.set_axis_off()
    azim, elev = 45, 19
    ax.view_init(azim=azim,elev=elev) 
    draw_BZ_border(ax, BZ_array, order_dir=np.array([azim,elev]), c='grey', alpha=0.8, lw=1)
   
   
    return fig, ax



### load data

kpts = np.loadtxt('kpts.dat')
bg = np.loadtxt('bg.dat')
phpop0 = np.loadtxt('phonon_pop_0ps.dat')[np.newaxis,:,:]
phpop1 = np.loadtxt('phonon_pop_1ps.dat')[np.newaxis,:,:]
phpop2 = np.loadtxt('phonon_pop_20ps.dat')[np.newaxis,:,:]
phpops = np.concatenate((phpop0, phpop1, phpop2), axis=0)
phpop_diff = phpops - phpops[0,:,:]


# vector of phonon mode numbers to plot
inuvec = [[0,1], [2], [4,5], [3], [0,1], [2], [4,5], [3]]


### plot

fig = plt.figure(figsize=(13,5), constrained_layout=True)
nrow = 2
ncol = 4
gs = gridspec.GridSpec(nrow, ncol, figure=fig, wspace=0.0, hspace=0.0)
colors = ['dodgerblue', 'dodgerblue', 'limegreen', 'orangered', 'gold', 'gold']
marker_size = 1 # this value is diffrent from the one plotted in the manuscript due to limitation of ipynb notebook
max_marker_size = 20
# the original setting on the manuscript is marker_size = 1, max_marker_size = 500

for iplot in range(nrow*ncol):
    it = iplot // ncol + 1
    ax = fig.add_subplot(gs[iplot], projection='3d')
    fig, ax = plot_border(fig, ax, kpts, bg) # add border 

    for inu in inuvec[iplot]:
   
        # make the plotted points larger
        phnorms = (phpop_diff[it,:,inu] ** 2.0) * marker_size
        newc = to_rgb(colors[inu])
        phnorms[phnorms > max_marker_size] = max_marker_size # set maximum marker size
        ax.scatter(kpts[:,0], kpts[:,1], kpts[:,2], s=phnorms,
                  marker='o', color=newc, alpha=0.3, lw=0)
        
    ax.set_box_aspect(None, zoom=1.6)
```
