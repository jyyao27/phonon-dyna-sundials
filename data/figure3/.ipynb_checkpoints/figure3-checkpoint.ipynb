```python
import numpy as np
import matplotlib.pyplot as plt
import colorsys

plt.rcParams['font.family'] = 'Arial'
plt.rcParams.update({'font.size':24, 'legend.fontsize':13})

### import all data
time = np.loadtxt('time_vec.dat')

# first column is electron error, second column is phonon error

rk4 = np.concatenate(
    (np.loadtxt('rk4-5.0.dat')[np.newaxis,:,:],
     np.loadtxt('rk4-0.5.dat')[np.newaxis,:,:],
     np.loadtxt('rk4-5.0-0.5.dat')[np.newaxis,:,:]))
erk = np.loadtxt('erk-5912.dat')
mri = np.concatenate((
    np.loadtxt('mri-5-5912.dat')[np.newaxis,:,:],
    np.loadtxt('mri-0.5-5912.dat')[np.newaxis,:,:]), axis=0)

# construct figure
fig, axes = plt.subplots(2, 1, figsize=(7, 10))
# set color and labels
rk4labels = [r'RK4, $h$ = 5 fs', r'RK4, $h$ = 0.5 fs',
             r'RK4, $h_s$ = 5 fs, $h_f$ = 0.5 fs']
erklabel = 'ERK'
mrilabels = [r'MRI, $h_s$ = 5 fs', r'MRI, $h_s$ = 0.5 fs']
markers = ['s', '.', '^']
markersizes = [8, 20, 10, 1, 8, 20]
num_rk4s = 3

rk4color = np.array([200, 103, 45]) / 255.0
erkcolor = [0.5,0.5,0.5]
mricolor = np.array([0., 142, 11.64]) / 255.0

# plot every 10th data
ir = np.arange(1,len(erk[:,0]), step=10)


for ip in range(2):
    
    i_markersize = 0
    
    # plot rk4 errors
    for im in range(3):
        axes[ip].plot(time, rk4[im][:,ip], label=rk4labels[im], color=rk4color, 
            marker=markers[im], markeredgecolor=rk4color, 
            markerfacecolor='white', markersize=markersizes[i_markersize], markevery=ir)
        i_markersize += 1
    
    # plot erk errors
    axes[ip].plot(time, erk[:,ip], label=erklabel, color=erkcolor, 
         marker='.', markeredgecolor=erkcolor, 
         markerfacecolor='white', markersize=markersizes[i_markersize], markevery=ir)
    i_markersize += 1
   
    # plot mri errors
    for im in range(2):
        axes[ip].plot(time, mri[im][:,ip], label=mrilabels[im], color=mricolor, 
            marker=markers[im], markeredgecolor=mricolor, 
            markerfacecolor='white', markersize=markersizes[i_markersize], markevery=ir)
  
        i_markersize += 1 

    axes[ip].set_yscale('log')
    axes[0].legend(bbox_to_anchor=(0.42,0.52),labelspacing=0.2)

axes[0].set_ylabel('Carrier population error')
axes[1].set_xlabel('Simulation time (ps)')
axes[1].set_ylabel('Phonon population error')
axes[0].xaxis.set_visible(False)
plt.tight_layout()
```
