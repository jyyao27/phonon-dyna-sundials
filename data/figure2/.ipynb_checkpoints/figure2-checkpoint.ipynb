```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import LogLocator

# load run time and error data, 
# first column is walltime, second column is carrier error, third column is phonon error
rk4 = np.loadtxt('rk4.dat')
erk = np.loadtxt('erk.dat')
mri = np.loadtxt('mri.dat')
data = {}
data['rk4'] = rk4
data['erk'] = erk
data['mri'] = mri

## Plot 
plt.rcParams.update({'font.size':24}) 
plt.rcParams['font.family'] = 'Arial'

label_vec = ['RK4','ERK','MRI']
marker_vec = ['s', '^', 'o']
color = np.array([[200,103,45],[125,125,125],[10,132,20]])/255.0
color_dark = np.where(color<0.8,color+0.2,1)

fig, ax = plt.subplots(1, 1, figsize=(10,8))

itemcount = 0
handles = []
for item in data.items():
    label = label_vec[itemcount]
    scat1 = ax.scatter(item[1][:,0], item[1][:,1], color=color_dark[itemcount], 
                       marker='o', s=300, edgecolors=color[itemcount], linewidths=3, label=label)
    scat2 = ax.scatter(item[1][:,0], item[1][:,2], color=color[itemcount], marker='x', s=200, lw=3)
    handles.append(scat1)
    itemcount +=1

# create legends for carriers and phonons
line1 = ax.scatter([], [], color=color[1], marker='x', s=200, linewidth=3, label='Carrier')
line2 = ax.scatter([], [], color=color_dark[1], marker='o', edgecolors=color[1], s=300, linewidth=3, label='Phonon')

ax.legend(handles=[line1, line2], loc='upper center', fontsize=24)

ax.set_xlabel('CPU core hours', fontsize=24)
ax.set_ylabel(r'Population error at $\it{t}$ = 0.5 ps', fontsize=24)

for text, i_color in zip(legend.get_texts(), color):
    text.set_color(i_color)  
legend = ax.legend(handles=handles, loc='best', fontsize=24)
ax.add_artist(legend)

# Adding the first legend manually to the plot
ax.set_ylim([0.3E-13,1E-1])
ax.text(5, 1E-9, 'decrease step size '+r'$h$', fontsize=22, color=color[0], rotation=-18)
ax.text(2, 5E-11, 'tighten tolerance rtol', fontsize=22, color=color[1], rotation=-20)
ax.text(0.2, 8E-14, 'decrease slow step size '+r'$h_s$', fontsize=22, color=color[2], rotation=-22)
ax.text(2.5, 4E-7, r'$h = 1$ fs', fontsize=22, color=color[0])
ax.text(150, 2E-9, r'$h = 0.01$ fs', fontsize=22, color=color[0])
ax.text(0.2, 3E-7, r'$h_{s}=20$ fs', fontsize=22, color=color[2])

# set label styles and ticks
ax.set_xscale('log')
ax.set_yscale('log')
customticks = [0.2,1,10,100]
customtick_labels = [0.2,1.0,10,100]
ax.set_xticks(customticks)
ax.set_xticklabels(customtick_labels)


plt.tight_layout()
```
